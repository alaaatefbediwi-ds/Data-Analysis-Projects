```python
# Customer Segmentation
df['Total_Sales_Per_Customer'] = df.groupby('Customer ID')['Sales'].transform('sum')
def segment_customer(sales):
    if sales < 1000:
        return 'Low Value'
    elif sales < 5000:
        return 'Medium Value'
    else:
        return 'High Value'

df['Customer_Value'] = df['Total_Sales_Per_Customer'].apply(segment_customer)
```

# Data Visualization

# 
-  Which product categories and sub-categories contribute the most to overall sales ?

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False).reset_index()
```

```python
# Total sales by Product Category
plt.figure(figsize=(12, 6))
sns.barplot(data=category_sales, x='Category', y='Sales', estimator='sum', ci=None)
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
grouped_data = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).reset_index()

# total sales by Product Sub-Category
plt.figure(figsize=(12, 6))
sns.barplot(data=grouped_data, x='Sub-Category', y='Sales', ci=None)
plt.title('Total Sales by Product Sub-Category')
plt.xlabel('Product Sub-Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
df.head(2)
```

# 
-  Are there certain regions or states that are significantly more profitable than others?

```python
sales_by_region = df.groupby('Region')['Sales'].sum().sort_values(ascending=False).reset_index()

# Total sales by region
plt.figure(figsize=(12, 6))
sns.barplot(data=sales_by_region, x='Region', y='Sales', ci=None)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
df.head(2)
```

# 
-  What are the top 5 products that have the highest sales volume?

```python
sales_by_product = df.groupby('Product Name')['Sales'].sum().reset_index()

# Sort by 'Sales' in descending order
sales_by_product_sorted = sales_by_product.sort_values(by='Sales', ascending=False).head(5)

# Plot total sales by Product
plt.figure(figsize=(15, 10))
sns.barplot(data=sales_by_product_sorted, x='Sales', y='Product Name', ci=None)
plt.title('Total Sales by Product')
plt.xlabel('Total Sales')
plt.ylabel('Product')
#plt.xticks(rotation=45)
plt.show()
```

#
-  Which customer segments are the most profitable?

```python
sales_by_segment = df.groupby('Segment')['Sales'].sum().sort_values(ascending=False).reset_index()

# Total profit by Customer Segments
plt.figure(figsize=(12, 6))
sns.barplot(data=sales_by_segment, x='Segment', y='Sales', ci=None)
plt.title('Total Sales by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
region_product_sales = df.groupby(['Region', 'Category'])['Sales'].sum().reset_index()

# Plot total sales by Product Category and Region
plt.figure(figsize=(14, 7))
sns.barplot(data=region_product_sales, x='Category', y='Sales', hue='Region', ci=None)
plt.title('Sales by Product Category and Region')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend(title='Region')
plt.show()
```

#
- Are there trends in customer preferences based on region?

```python
monthly_sales_by_region = df.groupby(['Ship Month', 'Region'])['Sales'].sum().reset_index()

# monthly sales trends by Region
plt.figure(figsize=(14, 7))
sns.lineplot(data=monthly_sales_by_region, x='Ship Month', y='Sales', hue='Region', marker='o')
plt.title('Monthly Sales Trends by Region')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend(title='Region')
plt.show()
```

#
- What are the months that have the maximum sales over each year?

```python
monthly_sales = df.groupby(['Ship Year', 'Ship Month'])['Sales'].sum().reset_index()

# Mapping months to month names
monthly_sales['Month_Name'] = monthly_sales['Ship Month'].apply(lambda x: pd.to_datetime(str(x), format='%m').strftime('%B'))

monthly_sales = monthly_sales.sort_values(by=['Ship Year', 'Ship Month'])

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

#total sales over months for each year
plt.figure(figsize=(14, 7))
sns.lineplot(data=monthly_sales, x='Month_Name', y='Sales', hue='Ship Year', marker='o', palette=colors)
plt.title('Total Sales Over Months for Each Year')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=avg_delivery_by_mode, x='Ship Mode', y='Delivery_Time', palette='Oranges_d')
plt.title('Average Delivery Time by Shipping Mode')
plt.xlabel('Shipping Mode')
plt.ylabel('Average Delivery Time (days)')
plt.show()
```

#
- Check whether high-value customers are more likely to make repeat purchases or not?

```python
retention_by_value = df.groupby('Customer_Value')['Order ID'].nunique().reset_index()

# Plot retention by customer value
plt.figure(figsize=(8, 5))
sns.barplot(data=retention_by_value, x='Customer_Value', y='Order ID', palette='Blues_d')
plt.title('Repeat Purchases by Customer Value')
plt.xlabel('Customer Value')
plt.ylabel('Number of Orders')
plt.show()
```
